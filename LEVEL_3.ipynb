```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("/content/Dataset .csv")
```

TASK - 3

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Relationship between price range, online delivery, and table booking
sns.boxplot(x=df['Price range'], y=df['Aggregate rating'], hue=df['Has Online delivery'])
plt.title("Price Range vs Rating with Online Delivery")
plt.show()

sns.boxplot(x=df['Price range'], y=df['Aggregate rating'], hue=df['Has Table booking'])
plt.title("Price Range vs Rating with Table Booking")
plt.show()
```

TASK - 2

```python
# Restaurants with highest and lowest votes
highest_votes = df.nlargest(5, 'Votes')[['Restaurant Name', 'Votes', 'Aggregate rating']]
lowest_votes = df.nsmallest(5, 'Votes')[['Restaurant Name', 'Votes', 'Aggregate rating']]

# Correlation between votes and rating
votes_rating_corr = df[['Votes', 'Aggregate rating']].corr()

print("Restaurants with Most Votes:\n", highest_votes)
print("Restaurants with Least Votes:\n", lowest_votes)
print("Correlation Between Votes and Ratings:\n", votes_rating_corr)
```
